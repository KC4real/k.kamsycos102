```python
import tkinter as tk
from tkinter import messagebox

# Function to calculate total order
def calculate_total_order():
    customer_name = entry_name.get()
    customer_order = entry_order.get()
    customer_menu = entry_menu.get()
    customer_quantity = int(entry_quantity.get())

    total_price = customer_quantity * menu_prices.get(customer_menu)

    if total_price < 2500:
        discount = total_price * 0.10
        total_price -= discount
        messagebox.showinfo("Total Price", f"Your total price after 10% discount: {total_price:.2f}")
    elif 2500 <= total_price < 5000:
        discount = total_price * 0.15
        total_price -= discount
        messagebox.showinfo("Total Price", f"Your total price after 15% discount: {total_price:.2f}")
    else:
        discount = total_price * 0.25
        total_price -= discount
        messagebox.showinfo("Total Price", f"Your total price after 25% discount: {total_price:.2f}")

# Create the GUI
root = tk.Tk()
root.geometry("500x300")
root.title("Pau Cafeteria Menu")

# Define menu prices
menu_prices = {"Jollof Rice":350,
               "Coconut Rice":350,
               "Jollof spaghetti":350,
               "Sweet chill chicken":1100,
               "Grilled chicken wigs":400,
               "Fried fish":500,
               "Boiled egg":200,
               "Salted sausages":200,
               "savoury beans":350,
               "Roasted sweet potatoes":300,
               "Fried plantain":150,
               "Mixed vegetable salad":150,
               "Boiled yam":150,
               "Eba":100,
               "Pounded yam":100,
               "semo":450,
               "Egusi soup":450,
               "water":200,
               "Glass drink":150,
               "PET Drink(35cl)":350,
               "Zobo drink":350,
               "Mango juice":350}

# Customer Information Labels and Entries
label_name = tk.Label(root, text="Enter Name:")
label_name.grid(row=0, column=0)
entry_name = tk.Entry(root)
entry_name.grid(row=0, column=1)

label_order = tk.Label(root, text="Enter Order:")
label_order.grid(row=1, column=0)
entry_order = tk.Entry(root)
entry_order.grid(row=1, column=1)

label_menu = tk.Label(root, text="Choose Menu:")
label_menu.grid(row=2, column=0)
entry_menu = tk.StringVar(root)
entry_menu.set("-------SELECT-MENU-------")  # Default value
menu_dropdown = tk.OptionMenu(root, entry_menu, *menu_prices.keys())
menu_dropdown.grid(row=2, column=1)

label_quantity = tk.Label(root, text="Enter Quantity:")
label_quantity.grid(row=3, column=0)
entry_quantity = tk.Entry(root)
entry_quantity.grid(row=3, column=1)

# Calculate Button
calculate_button = tk.Button(root, text="Calculate", command=calculate_total_order)
calculate_button.grid(row=4, column=0, columnspan=2)

root.mainloop()
```

```python
1.create a tkinter window
2. set a desired size and title
3.create a desired menu with the prices
4.Create labels and entry fields for customer information
5. to get the menu selection go to optionMenu
6.define calculate_total_order() function
7. from the the menu calculate the total order
8. apply the discounts for each price range
9.create a button that will launch the calculation
10.Start the tkinter event loop
```
